```python
class Detector(object):
    """
    A Detector object characterises a gravitational wave (GW) interferometric detector
    """

    def __init__(self, detector):
        """
        detector  -- label string of the detector
        descriptor -- LAL descriptor
        location -- geographic location of the detector
        response -- response matrix

        """
        self.name = detector
        self.descriptor =  CachedDetectors[DETECTOR_SITES[detector]]
        self.location = lalsimulation.DetectorPrefixToLALDetector(detector).location
        self.response = lalsimulation.DetectorPrefixToLALDetector(detector).response
        
    def antenna_pattern(self, time_at_coalescence, RA, dec,psi):
        """ Compute antenna response
        """
        
        gmst_rad = lal.GreenwichMeanSiderealTime(time_at_coalescence)
        fplus = []
        fcross = []
    
        for (RA_val, dec_val) in zip(RA, dec):
            fplus_val,fcross_val = ComputeDetAMResponse(self.descriptor.response,
                                    RA_val, dec_val, psi, gmst_rad)
            fplus.append(fplus_val)
            fcross.append(fcross_val)
        return np.array(fplus), np.array(fcross)
```

```python
def generateAntennaPatterns(time_0, ra_plot, dec_plot,psi):

    #Generate Grid for the calculation
    ra, dec = np.meshgrid(ra_plot,dec_plot)

    for i in range(4):    
        if(i==0):
            detector = Detector('H1')
        if(i==1):
            detector = Detector('L1')
        if(i==2):
            detector = Detector('V1')
        if(i==3):
            detector = Detector('K1')

        Fplus,Fcross = detector.antenna_pattern(time_0, ra.flatten(), dec.flatten(),psi)

        if(i==0):
            FplusH1 = Fplus.reshape(ra.shape)
            FcrossH1 = Fcross.reshape(ra.shape)
        if(i==1):
            FplusL1 = Fplus.reshape(ra.shape)
            FcrossL1 = Fcross.reshape(ra.shape)
        if(i==2):
            FplusV1 = Fplus.reshape(ra.shape)
            FcrossV1 = Fcross.reshape(ra.shape)
        if(i==3):
            FplusK1 = Fplus.reshape(ra.shape)
            FcrossK1 = Fcross.reshape(ra.shape)
            
    return FplusH1, FcrossH1, FplusL1, FcrossL1, FplusV1, FcrossV1, FplusK1, FcrossK1
    
```

```python
def standarddev(x,y):
    dx = x[1]-x[0]
    normalize = np.sum(y)*dx
    mean = np.sum(x*y)*dx
    std = np.sqrt(np.sum((x-mean)**2*y)*dx/normalize)
    return std

def median(x,y):
    rv = st.rv_discrete(values=(x,y/(np.sum(y))))
    return rv.median()


def skyplot(x, ra_plot = None, dec_plot = None, save = False):    
    n = len(x)
    if(ra_plot is None):
        ra_plot = np.linspace(-np.pi,np.pi, n)
        dec_plot= np.linspace(-np.pi/2, np.pi/2, n)

    ra_mesh, dec_mesh = np.meshgrid(ra_plot,dec_plot)

    fig = plt.figure(1)
    ax = fig.add_subplot(111, projection='mollweide')
    fig.set_size_inches(12.5, 7.5)
    plt.pcolormesh(ra_mesh, dec_mesh,x, cmap = cm.jet)
    plt.colorbar()
    plt.grid(alpha=0.3)
    plt.title('$\Delta D/D$ for HLVK, $v=$ {}, $\psi=$ {:.4f}, sens: {} for t = {} s'.format(v_0, psi_init, sensitivity, time_0))
    if(save):
        plt.savefig('figures/version3stdev_rel_distance_corr_iota{}resolution{}psi{}sens{}.png'.format(v_0,n,psi_init, sensitivity), dpi = 200)
    plt.show()
    
#Convert coordinates to indizes
def convert(ra, dec, n):
    ra = np.array(ra)
    dec = np.array(dec)
    indra = n*(np.pi+float(ra))/(2*np.pi)
    if(indra>n):
        indra = indra - n
    inddec = n*(np.pi/2+float(dec))/np.pi
        
    return int(indra), int(inddec)

grav_constant = 6.674*10**(-11)
c = 3*10**8
m_chirp = 1.1975*2*10**30
Mpc = 3.086*10**22

def calculate_distance(m_chirp, epsilon, sigma_d, delta_psi):
    kscalark = 5/6*np.pi**(-4/3)*(grav_constant*m_chirp)**(5/3)*c**(-3)*np.sum(freqs**(-7/3)*PSD_average_inverse*df)
    
    distance_val = np.sqrt(kscalark)/SNRfix*np.sqrt(sigma_d*((1+v_0**2)**2/4+v_0**2+epsilon*(1-v_0**2)**2/4*np.cos(4*(psi_init+delta_psi))))
    return distance_val/Mpc

def SNRatHLVK(time_0, ra_plot, dec_plot,v_val, distance_val):
    "Computes the SNR at each detector"
    
    n_ra = len(ra_plot)
    n_dec = len(dec_plot)
    
    
    FplusH1, FcrossH1, FplusL1, FcrossL1, FplusV1, FcrossV1, FplusK1, FcrossK1 = generateAntennaPatterns(time_0, ra_plot, dec_plot, psi_init)

    kscalark = np.zeros(n_detec)
    SNRsq = np.zeros((n_dec, n_ra, n_detec))

    for i in range(n_detec):
        kscalark[i] = 5/6*np.pi**(-4/3)*(grav_constant*m_chirp)**(5/3)*c**(-3)*np.sum(freqs**(-7/3)*PSD[:,i]**(-1)*df)

    SNRsq[:,:,0] = (FplusH1**2*((1+v_val**2)/2)**2+FcrossH1**2*v_val**2)*kscalark[0]/(distance_val*Mpc)**2
    SNRsq[:,:,1] = (FplusL1**2*((1+v_val**2)/2)**2+FcrossL1**2*v_val**2)*kscalark[1]/(distance_val*Mpc)**2
    SNRsq[:,:,2] = (FplusV1**2*((1+v_val**2)/2)**2+FcrossV1**2*v_val**2)*kscalark[2]/(distance_val*Mpc)**2
    if(n_detec==4):
        SNRsq[:,:,3] = (FplusK1**2*((1+v_val**2)/2)**2+FcrossK1**2*v_val**2)*kscalark[3]/(distance_val*Mpc)**2
    
    return np.sqrt(SNRsq)
```

```python
def findlocalextremum(mat, limits = None, use_min = True):
    ''' Result the row and col indices of local extremum in 
        a submatrix defined by limits = [xmin, xmax, ymin, ymax]
    '''
        
    extremum_func = np.argmin if use_min else np.argmax
    if limits is None:
        limits = [0, mat.shape[0], 0, mat.shape[1]]
    sub = mat[limits[0]:limits[1],limits[2]:limits[3]]
    offset = np.array([limits[0], limits[2]])
    return tuple(np.unravel_index(extremum_func(sub, axis=None), sub.shape) + offset)
```

```python
def ThetaFunctionAnalytic(ra_plot, dec_plot):
    opt_event_pos = []
    n_ra = len(ra_plot)
    n_dec = len(dec_plot)

    FplusH1, FcrossH1, FplusL1, FcrossL1, FplusV1, FcrossV1, FplusK1, FcrossK1 = generateAntennaPatterns(time_0, ra_plot,  dec_plot,0)
    
    Theta = np.zeros([2,2,n_dec,n_ra]) 
    Theta_diag = np.zeros([n_dec,n_ra])
    epsilon = np.zeros([n_dec,n_ra])
    sigma_d = np.zeros([n_dec,n_ra])

    for i in range(n_ra):
        for j in range(n_dec):
            Theta[0,0,j,i] = kappa[0]*FplusH1[j,i]**2 + kappa[1]*FplusL1[j,i]**2 + \
                                kappa[2]*FplusV1[j,i]**2 + kappa[3]*FplusK1[j,i]**2
            Theta[0,1,j,i] = kappa[0]*FplusH1[j,i]*FcrossH1[j,i] + kappa[1]*FplusL1[j,i]*FcrossL1[j,i] + \
                            + kappa[2]*FplusV1[j,i]*FcrossV1[j,i]+kappa[3]*FplusK1[j,i]*FcrossK1[j,i]
            Theta[1,0,j,i] = Theta[0,1,j,i]
            Theta[1,1,j,i] = kappa[0]*FcrossH1[j,i]**2 + kappa[1]*FcrossL1[j,i]**2 +\
                            kappa[2]*FcrossV1[j,i]**2 +kappa[3]*FcrossK1[j,i]**2

    delta_psi = np.arctan(2*Theta[0,1]/(-Theta[1,1]+Theta[0,0]))/4.
    c, s = np.cos(2*delta_psi), np.sin(2*delta_psi)
    R = np.array([[c,s],[-s, c]])
    Rtranspose = np.array([[c,-s],[s, c]])

    ThetaDiag = np.zeros([2,2,n_dec,n_ra])

    for i in range(n_dec):
        for j in range(n_ra):
            ThetaDiag[:,:,i,j] = np.dot(R[:,:,i,j],np.dot(Theta[:,:,i,j],Rtranspose[:,:,i,j]))
            
    epsilon = (ThetaDiag[0,0,:,:]-ThetaDiag[1,1,:,:])/(ThetaDiag[0,0,:,:] + ThetaDiag[1,1,:,:])
    
    for i in range(n_dec):
        for j in range(n_ra):
            if(epsilon[i,j]<0):
                epsilon[i,j] = -epsilon[i,j]
                delta_psi[i,j] = delta_psi[i,j] - np.pi/4

    sigma_d = 0.5*(Theta[0,0,:,:]+Theta[1,1,:,:])
    
    distance_val = calculate_distance(m_chirp, epsilon, sigma_d, delta_psi)
    SNR_val = SNRatHLVK(time_0, ra_plot, dec_plot,v_0, distance_val)
    
    if((n_ra+n_dec)==2):
        print ('distance = {}'.format(distance_val))
        print (SNR_val)

    return epsilon, sigma_d, Theta, delta_psi, distance_val, SNR_val
```

```python
def DistanceError(ra_point,dec_point,corr=True, n_g = None, plot = False):
    "Computes the distance inclination posterior, returns the std over the median distance, the distance posterior,"
    "the distance to the event and the SNR in each detector"
    
    n_plot_ra, n_plot_dec = len(ra_point), len(dec_point)
    epsilon_val, sigma_d_val, _, delta_psi_val, distance_val, SNR_val = ThetaFunctionAnalytic(ra_point, dec_point)

    D_max = 3.5

    rho_0_val = SNRfix/np.sqrt(sigma_d_val*((1+v_0**2)**2/4+v_0**2+epsilon_val*(1-v_0**2)**2/4*np.cos(4*(psi_init+delta_psi_val))))  
    
    #number of grid points
    if(n_g is None):    
        n_g = 2000

    #grid
    v = np.linspace(-1,1,n_g)
    D_rel = np.linspace(0+1e-5,2*D_max,n_g)

    delta_x = D_rel[1]-D_rel[0]

    v_int,D_rel_val = np.meshgrid(v,D_rel)

    delta1 = np.zeros((n_plot_dec,n_plot_ra))
    delta2 = np.zeros((n_plot_dec,n_plot_ra))
    correction = np.zeros((n_plot_dec,n_plot_ra))

    PDF_D_rel_incl = np.zeros((n_plot_dec,n_plot_ra,n_g, n_g) )
    PDF_D_rel_incl_norm = np.zeros((n_plot_dec,n_plot_ra,n_g, n_g) )
    PDF_D_rel = np.zeros((n_plot_dec,n_plot_ra, n_g) )

    D_rel_mean = np.zeros((n_plot_dec,n_plot_ra))
    sigmasq_discrete = np.zeros((n_plot_dec,n_plot_ra))
    stdev_rel_distance = np.zeros((n_plot_dec,n_plot_ra))
    medianD = np.zeros((n_plot_dec,n_plot_ra))

    if(corr==False):
        delta1 = 1/rho_0_val*np.sqrt((1+epsilon_val*np.cos(4*(psi_init+delta_psi_val)))/(sigma_d_val*(1-epsilon_val**2)))
        delta2 = 1/rho_0_val*np.sqrt((1-epsilon_val*np.cos(4*(psi_init+delta_psi_val)))/(sigma_d_val*(1-epsilon_val**2)))
    
    if((len(ra_point)+len(dec_point))==2):  
        print('epsilon = {:.4f}, sigma = {:.4f}, psi_bar = {:.4f}'.format(float(epsilon_val), float(sigma_d_val), psi_init + float(delta_psi_val)))
        print('rho_0_val = {}'.format(float(rho_0_val)))
        if(corr==False):
            print('delta 1 = {}, delta2 = {}'.format(delta1, delta2)) 
    
    for i in range(n_plot_ra):
        for j in range(n_plot_dec):
            #print delta1[i,j], delta2[i,j]
            if(corr):
                PDF_D_rel_incl[j,i,:,:] = D_rel_val**2*np.e**(-1/2*rho_0_val[j,i]**2*sigma_d_val[j,i]*((1-epsilon_val[j,i]*\
                                        np.cos(4*(psi_init+delta_psi_val[j,i])))*(v_int/D_rel_val-v_0)**2+(1+epsilon_val[j,i]*\
                                        np.cos(4*(psi_init+delta_psi_val[j,i])))*((1+v_int**2)/(2*D_rel_val)-(1+v_0**2)/2)**2))
            else:
                PDF_D_rel_incl[j,i,:,:] = D_rel_val**2*np.e**((-1/(2*delta1[j,i]**2)*(v_int/D_rel_val-v_0)**2-1/(2*delta2[j,i]**2)*((1+v_int**2)/(2*D_rel_val)-(1+v_0**2)/2)**2))
                
            
            
            PDF_D_rel_incl_norm[j,i,:,:] = PDF_D_rel_incl[j,i,:,:]/(np.sum(PDF_D_rel_incl[j,i,:,:]))

            PDF_D_rel[j,i,:] = np.sum(PDF_D_rel_incl_norm[j,i,:,:] ,axis = 1)/delta_x
            D_rel_mean[j,i] = np.sum(D_rel*PDF_D_rel[j,i])*delta_x
            #sigmasq_discrete[j,i] = np.sum((D_rel-D_rel_mean[j,i])**2*PDF_D_rel[j,i])*delta_x

            stdev_rel_distance[j,i] = standarddev(D_rel, PDF_D_rel[j,i,:])
            medianD[j,i] = median(D_rel,PDF_D_rel[j,i,:])

    deltaDoverD = np.array(stdev_rel_distance/medianD)
        #print('DeltaD/D = {}'.format(deltaDoverD))
    if(plot):
        plt.plot(D_rel, PDF_D_rel[j,i,:], label = '{}'.format(v_0))
        plt.xlim(0.2,2)
        plt.ylim(0)
        plt.legend()
        #plt.savefig('varyinginclination.png', dpi = 200)
        plt.show

        plt.figure(figsize=(14,7))
        im = plt.pcolormesh(np.arccos(v_int), D_rel_val, PDF_D_rel_incl_norm[0,0], cmap = cm.jet)
        plt.colorbar(im)

        plt.title('PDF $D_{rel}$ inclination')
        plt.xlabel('$ \iota$')
        plt.ylabel('$Distance$')
        #plt.xlim(-1,0)
        plt.ylim(0,2)

        plt.show()
        
    if(corr):
        np.savetxt('distance_v_{}_psi_{}_corr.txt'.format(v_0,psi_init), np.column_stack((D_rel,PDF_D_rel[0,0,:])))
    
    return deltaDoverD, PDF_D_rel, distance_val, SNR_val
    '''
    else:
        np.savetxt('HLV error skymaps psi 0/stdev_rel_distance_{}.txt'.format(v_0), stdev_rel_distance)    '''    
```
